## 4. Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(8,9))
sns.histplot(data=df,x="GPA",bins=20,kde=True,color="skyblue")
plt.title("Distribution of GPA")
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.savefig("../images/gpa_histogram.png",dpi=300,bbox_inches="tight")

plt.show()
```

### Box Plot of GPA

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df,x="GPA",color="lightgreen")
plt.title("box plot of gpa")
plt.xlabel("GPA")
plt.savefig("../images/gpa_boxplot.png",dpi=300, bbox_inches="tight")
plt.show()
```

### Gender Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x="Gender")
plt.title("Distribution of Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.savefig("../images/gender_countplot.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Relationship between Study Time and GPA

```python
plt.figure(figsize=(8,5))
sns.scatterplot( data=df, x="StudyTimeWeekly", y="GPA")
plt.title("Study Time vs GPA")
plt.xlabel("Study Time Weekly (Hours)")
plt.ylabel("GPA")
plt.savefig("../images/Studytime_vs_gpa.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap( df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.savefig("../images/correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Pair Plot of Numerical Features

```python

sns.pairplot(df[["GPA","StudyTimeWeekly","Absences","Age"]],diag_kind="kde")
plt.savefig("../images/pairplot.png", dpi=300)
plt.show()
```

## 5. Key Insights

- The dataset contains no missing values.
- The dataset contains no duplicate records.
- Most students have GPA between 1.5 and 3.0.
- GPA does not contain significant outliers.
- Gender distribution is almost balanced.
- Absences have a strong negative correlation with GPA.
- StudyTimeWeekly has a weak positive relationship with GPA.
- Pair plot confirms the relationship between the numerical variables.
